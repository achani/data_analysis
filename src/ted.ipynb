## 5. What were the best events (not talk)  in TED history to attend. 

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
ted['ratings'][0]
```

```python
positive = ['Funny', 'Beautiful', 'Ingenious', 'Courageous', 'Informative', 'Fascinating', 'Persuasive', 'Jaw-dropping', 'Inspiring']
negative = ['Longwinded', 'Confusing','Unconvincing', 'OK', 'Obnoxious']
import json 
def get_score(raw):
    raw = raw.replace("\'", "\"")
    data = json.loads(raw)
    plus = sum([int(x['count']) for x in data if x['name'] in positive])
    minus = sum([int(x['count']) for x in data if x['name'] in negative])
    return plus - minus

ted['score'] = ted['ratings'].apply(get_score)
    
```

```python
ted['score']
```

## 8. Which occupations deliver the funniest TED Talks on average? 

### Approach 1 (not very good because we don't consider how many total ratings are there)

```python
ted['funny_rating'] = ted['ratings_list'].apply(lambda x: sum([i['count'] for i in x if i['name'] == "Funny"]))
```

```python
ted.groupby("speaker_occupation").funny_rating.mean().sort_values(ascending=False)
#ted.columns
```

### Approach 2: Create a col for what percentage of total ratings are funny 

This approach also has the problem that the sample size in some case is too small. there are many occupations where there is only one speaker. 

```python
ted['total_ratings'] = ted.ratings_list.apply(lambda x: sum([i['count'] for i in x]))
```

```python
ted['funny_pct'] = ted['funny_rating'] / ted['total_ratings']
```

```python
ted.head()
```

```python
ted.groupby("speaker_occupation").funny_pct.mean().sort_values(ascending=False)
```

### Approach 3.1 - Focus on the occupations that are relatively well represented

```python
occupation_counts = ted.speaker_occupation.value_counts()
```

```python
top_occupations = occupation_counts[occupation_counts >= 5].index 
```

```python
ted_top_occupations = ted[ted.speaker_occupation.isin(top_occupations)]
```

```python
ted_top_occupations.shape
```

```python
ted_top_occupations.groupby("speaker_occupation").funny_pct.mean().sort_values(ascending=False)
```
