### Filtering data - Using Conditionals 


```python
#Filter mask is returned by following line. True where condition is met and Falst otherwise. Return Type : Series (containing True or False)
df_temp['last'] == "Chani"
```

```python
filt = (df_temp['last'] == 'Chani')
```

```python
df_temp[filt]
```

```python
#Other way to do the same. Benefit of using this method is that you can still select the Columns you need. 
df_temp.loc[filt, 'email']
```

```python
# and and or operator (Cannot use the same symbol as Python. So need to use & for and AND | for or 
# (doesn't work without proper parenthesis) filt = df_temp['email'].str.contains('mail') & df_temp['last'] == "Chani"
filt = (df_temp['email'].str.contains('mail')) | (df_temp['last'] == "Patil")
```

```python
df_temp.loc[filt]
```

```python
# '~' Negates the filter
filt = (df_temp['email'].str.contains('mail')) & (df_temp['last'] == "Chani")
```

```python
df_temp.loc[~filt]
```

### apply method for Series objects

```python
#apply is used for calling a function on the values. works for DataFrame or a Series
#apply function to all values in a Series

df_temp['email'].apply(len) #get length of the email for all values 
```

### Apply method for DataFrame objects
For a dataframe the apply runs a function on each row and column of the DataFrame


```python
#The following gives unexpected output. We might expect it will apply the function to each values of each row. 
#However it applies to each Series (specifically each Column) 

### 

# index    3 (index has 3 values)
# email    3 
# first    3
# last     3
# dtype: int64

###
df_temp.apply(len)
```

```python
#We can do the same for a specific column
len(df_temp['email'])
```

```python
#We can specify axis (i.e. apply based on columns instead of default 'rows')
# in this case it shows how many columns in each row. 
df_temp.apply(len, axis='columns')
```

```python
#other functions that can be used for apply 
df_temp.apply(pd.Series.min)
```

```python
#Can also use lambda functions (works on a Series) 
df_temp.apply(lambda x: x.min())
```

## What if we need to know what %age of people from a country know Python. 

```python
#What if we need to know what %age of people from a country know Python. 
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python

python_df = pd.concat([country_respondents,country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'count': 'NoOfRespondants', 'LanguageWorkedWith': 'NoKnowsPython'}, inplace=True)
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NoKnowsPython']/python_df['NoOfRespondants']*100)
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
