## Tutorial 2 
### DataFrame and Series


```python
people = {
    "first" : ["Ajay" , "Nitin" , "Raj" ],
    "last" : ["Chani", "Naraaz", "Patil"],
    "email" : ["ajay@gmail.com", "nitin@hotmail.com" , "raj@ms.com" ]
}
```

```python
#we can create a DF from a dictionary
df_temp = pd.DataFrame(people)
```

```python
df_temp
```

```python
#Access a column just like a key in a dictionary
df_temp['email']
```

```python
#Type of the column is "pandas.core.series.Series" => a One Dimensional Array
type(df_temp['email'])
#can also use "." but better to use first method as . notation may cause errors like column name clashing with a method of DataFrame.
df_temp.email 
```

```python
#Select multiple columns. When selecting multiple columns, it is no longer a Series. 
#It returns a new DataFrame now. Index here will be integers 0,1,2.. 
df_temp[['last', 'email']]
```

```python
#Get a list of all columns
df_temp.columns 
```

```python
#To get rows we can use "loc" or "iloc" indexes 
#iloc => access rows by Integer location

#to get first row. Returns a series containing values in the first row. Index here will be column names. 
df_temp.iloc[0]
```

```python
#Select multiple rows and specific columns. Cannot use column name. must use indexes.

df_temp.iloc[[0,1], [0,1]]
```

```python
# 'loc' is to select by "labels" 
df_temp.loc[0] #here 0 is used as label becuase the index has values 0,1,2..
```

```python
#we can now use column name.
df_temp.loc[[0,1],['email','first']]
```
